```python
import pandas as pd

data = {"roll_no": [101,102,103,104,105],
        "gender":["female","male","female","male","female"]
        ,"department":["computer","it","entc","mechanical","computer"],
        "CGPA":[8.2,7.5,9.1,6.8,8.7]
 }
df = pd.DataFrame(data)
df.to_csv("students.csv",index=False)
df
```

```python
df.shape
```

```python
#access the informarion about Department column
df["department"]
```

```python
#acess infomation about prn 103
df[df['roll_no'] == 103]
```

```python
#perfrom mean,medin and mode of the CGPA
print("Mean of CGPA is: ",df['CGPA'].mean())
print("Median of CGPA is: ",df['CGPA'].median())
print("Mode of CGPA is: ",df['CGPA'].mode())
```

```python
#find our data type of each column
df.dtypes
```
